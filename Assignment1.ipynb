#No 2, 3, 4, 5


```python
#  BMI Calculation with Status
df_g['BMI'] = df_g['Berat Badan (Kg)'] / ((df_g['Tinggi Badan (Cm)'] / 100)**2)

# Nomor Membuat kolom Status BMI
def status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Unknown'

df_g['Status BMI'] = df_g['BMI'].apply(status_bmi)
df_g.head(25)
```

```python
# Nomor 9: Pencarian Nilai
# A. Cari nilai terendah dan tertinggi
height_min = df_g['Tinggi Badan (Cm)'].min()
height_max = df_g['Tinggi Badan (Cm)'].max()
weight_min = df_g['Berat Badan (Kg)'].min()
weight_max = df_g['Berat Badan (Kg)'].max()

# Cari nilai terendah dan tertinggi beserta Nama siswa
height_min_name = df_g.loc[df_g['Tinggi Badan (Cm)'].idxmin()]['Nama']
height_max_name = df_g.loc[df_g['Tinggi Badan (Cm)'].idxmax()]['Nama']

weight_min_name = df_g.loc[df_g['Berat Badan (Kg)'].idxmin()]['Nama']
weight_max_name = df_g.loc[df_g['Berat Badan (Kg)'].idxmax()]['Nama']

print('Nilai Terendah Tinggi Badan:', height_min, 'Cm (', height_min_name, ')')
print('Nilai Tertinggi Tinggi Badan:', height_max, 'Cm (', height_max_name, ')')
print('Nilai Terendah Berat Badan:', weight_min, 'Kg (', weight_min_name, ')')
print('Nilai Tertinggi Berat Badan:', weight_max, 'Kg (', weight_max_name, ')')
```
